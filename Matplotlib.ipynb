```python
##Which one should you use (pyplot vs matplotlib oo method)
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x="age",y="chol",c="target");
```

```python
# OO method
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",y="chol",c="target",
            ax=ax);
ax.set_xlim([45,100])
```

```python
# OO method from scratch
fig,ax=plt.subplots(figsize=(10,6),sharex=True)
# Plot the data

scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);
# Customize the plot
ax.set(title="heart disease and cholesterol levels",
      xlabel="Age",ylabel="Cholesterol");
# Add a legend
ax.legend(*scatter.legend_elements(),title="Target");
# add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
over_50.head()
```

```python
#subplot od chol,age,thalach
fig,(ax0,ax1)=plt.subplots(figsize=(10,10) ,nrows=2,ncols=1,sharex=True)

#add data 
scatter=ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
#customizing the ax0
ax0.set(title="Heart disease and cholestrol level",ylabel="cholestrol")
#add legend
ax0.legend(*scatter.legend_elements(),title="Target")
#add a mean line
ax0.axhline(y=over_50["chol"].mean(),linestyle="--")


#Add Data to ax1
scatter1=ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])
#customize 
ax1.set(title="heart disease and thalach level",xlabel="age",ylabel="thalach");
#add a legent
ax1.legend(*scatter1.legend_elements(),title="Target");
# Adding line
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");

#add a title to the figure
fig.suptitle("Heart disease analysis",fontsize=16,fontweight="bold")
```
