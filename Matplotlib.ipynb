```python
x=[1,2,3,4]
y=[11,22,33,44]
plt.plot(x,y)
```

```python
# first method
fig=plt.figure()# creates a figure
ax=fig.add_subplot()#add some axes
plt.show()
```

```python
# 2nd method
fig=plt.figure()#creates a figure
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# 3rd method recommended
fig,ax=plt.subplots()
ax.plot(x,y)# add some plots
```

```python
# matplotli exampple workflow


%matplotlib inline
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[11,22,33,44]

# Setup plot

fig, ax=plt.subplots(figsize=(10,10))
ax.plot(x,y)

# plot data
ax.plot(x,y)

# Customize plot
ax.set(title="simple plot",
      xlabel="x-axix",
      ylabel="y-axis")
# save and show
fig.savefig("images/sample-plot.png")
```
